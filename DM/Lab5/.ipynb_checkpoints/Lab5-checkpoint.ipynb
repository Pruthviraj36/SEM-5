### 1)  First, you need to read the titanic dataset from local disk and display Last five records



```python
import pandas as pd

df = pd.read_csv('titanic.csv')

df.shape
```

```python
df.tail(5)
```

### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
df.isnull().sum()

dfclean = df.dropna(thresh=df.shape[1] * 0.8, axis=0).copy()

dfclean.shape
```

```python
dfclean['Age'] = dfclean['Age'].fillna(dfclean['Age'].median())
dfclean['Embarked'] = dfclean['Embarked'].fillna(dfclean['Embarked'].mode()[0])
dfclean['Fare'] = dfclean['Fare'].interpolate()
# dfclean['Cabin'] = dfclean['Cabin'].interpolate()

dfclean
```

```python
dfclean.isnull().sum()
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
dfclean['min_max_of_age'] = (dfclean['Age'] - dfclean['Age'].min())/(dfclean['Age'].max() - dfclean['Age'].min())

dfclean.decimal_scaling_of_age
```

```python
dfclean['Z_score_scaling_of_age'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()

dfclean.Z_score_scaling_of_age
```
