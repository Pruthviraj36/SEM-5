### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
quantitative_cols = df.select_dtypes(include='number').columns

for col in quantitative_cols:
    data = df[col].dropna()
    mean = round(data.mean(), 2)
    std = round(data.std(), 2)
    min_val = round(data.min(), 2)
    max_val = round(data.max(), 2)
    val_range = round(max_val - min_val, 2)
    mode_val = data.mode().iloc[0] if not data.mode().empty else 'N/A'

    print(f"{col}:")
    print(f"\t Mean = {mean}")
    print(f"\t Standard deviation = {std}")
    print(f"\t Minimum = {min_val}")
    print(f"\t Maximum = {max_val}")
    print(f"\t Range = {val_range}")
    print(f"\t Mode = {mode_val}")
    print()
```

```python
quantitative_df = df.select_dtypes(include='number')


stats = pd.DataFrame({
    'Mean': quantitative_df.mean(),
    'Std Dev': quantitative_df.std(),
    'Min': quantitative_df.min(),
    'Mode': quantitative_df.mode().iloc[0],
    'Range': quantitative_df.max() - quantitative_df.min(),
    'Max': quantitative_df.max()
})
print(stats)
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt
```

```python
age_data = df['Age'].dropna()

plt.figure(figsize=(8, 5))
plt.hist(age_data, bins=8, edgecolor='black')
plt.title('Histogram of Age (8 Bins)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
numeric_cols = df.select_dtypes(include='number')

numeric_cols.boxplot()
plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
df = df[['Age', 'Fare', 'SibSp', 'Parch', 'Pclass']].dropna()

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.scatter(df['Age'], df['Fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.subplot(2, 3, 2)
plt.scatter(df['Age'], df['SibSp'])
plt.title('Age vs SibSp')
plt.xlabel('Age')
plt.ylabel('SibSp')

plt.subplot(2, 3, 3)
plt.scatter(df['Age'], df['Parch'])
plt.title('Age vs Parch')
plt.xlabel('Age')
plt.ylabel('Parch')

plt.subplot(2, 3, 4)
plt.scatter(df['Fare'], df['Pclass'])
plt.title('Fare vs Pclass')
plt.xlabel('Fare')
plt.ylabel('Pclass')

plt.subplot(2, 3, 5)
plt.scatter(df['SibSp'], df['Parch'])
plt.title('SibSp vs Parch')
plt.xlabel('SibSp')
plt.ylabel('Parch')

plt.tight_layout()
plt.show()
```
